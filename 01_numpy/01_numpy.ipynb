##### memory requirement numpy vs list

```python
import numpy as np

li_arr=[i for i in range(100)]
np_arr=np.arange(100)

print(li_arr)
print(np_arr)
```

```python
#size of one element in numpy array
print(np_arr.itemsize)

#size of 100 element in numpy array
print(np_arr.itemsize * np_arr.size)
```

```python
import sys

#size of 1 elem in list
print(sys.getsizeof(li_arr[0]))

#size of 100 ele in li
print(sys.getsizeof(1)*len(li_arr))
```

##### numpy vs list speed by doing addition

```python
import time
import numpy as np

size=100000

def add_using_list():
    t1=time.time()
    a=range(size)
    b=range(size)
    c=[a[i]+a[i] for i in range(size)]
    t2=time.time()
    return (t2-t1)

def add_using_numpy():
    t1=time.time()
    a=np.arange(size)
    b=np.arange(size)
    c=a+b
    
    t2=time.time()
    return (t2-t1)

t_l=add_using_list()
np_t=add_using_numpy()
print(t_l*1000,np_t*1000)  #millisecondsd
```

##### creating numpy arrays using n.array(array_like_object,dtype)

```python
# using n.array(array_like_object,dtype)

a=[1,2,3]
b=np.array(a)
print(type(b))
c=np.array("abc")   #consider string as 1 unit
print(c)

d=np.array((1,2,3))   #tuple is ok
print(d)
```

```python
a=[1,2,3,'a',5]
b=np.array(a)
print(b)     #numpy contain homogenous elements
```

```python
a=[1,2,3,'a',5]
b=np.array(a,dtype=int)
print(b)
```

```python
a=[1,2,3,4,5]
b=np.array(a,dtype=int)
print(b)
```

```python
a=[1,2,3,4,5]
b=np.array(a*3)  #here mul list with int
print(b)
print(b*3)  #in case of numpy each elemnt will mul wit 3
```

```python
# creating numpy with each val=1......np.ones(dimension,dtype)

a=np.ones(5)
b=np.ones((2,3),dtype=int)
print(a)
print(b)
```

```python
# creating numpy with each val=0......np.zeros(dimension,dtype)

a=np.zeros(5)
b=np.zeros((2,3),dtype=int)
print(a)
print(b)
```

```python
# creating numpy with each val=0......np.full(dimension,fill_value,dtype)

a=np.full(5,fill_value=22)
b=np.full((2,3),2)
print(a)
print(b)
```

```python
# creating numpy with each val=1......np.empty(dimension,dtype)
#np.empty fill garbage values which afre already present in memory----faster than np.zeros or ones

a=np.empty(5)
b=np.empty((2,3),dtype=int)
print(a)
print(b)
```

```python
#np.arange(start,end,step,dtype).....like range()
a=np.arange(5,29,2)
print(a)
```

```python
#np.linspace(start,end,np_ofvalues_you_want,dtype)
a=np.linspace(5,29,5)  #both start and step are included
print(a)
```

```python
a=np.linspace(5,29,10)
print(a)
common_diff=a[1]-a[0]
print(common_diff)
```

```python
a=np.linspace(5,29,6,endpoint=False) #end will be excluded
print(a)
```

```python
a=np.identity(3)  #take single arg--always generate square matrix
print(a)
```

```python
a=np.eye(3) 
print(a)
```

```python
a=np.eye(3,4)  #can take m,n as arg
print(a)
```
