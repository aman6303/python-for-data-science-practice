```python
#work on specific part of an 2d arr

print(x)
```

```python
#let update value of 100 to 99 in column 4
boolar= x[:,4]==100
boolar
```

```python
x[boolar,4]=99   #boolar and column -row taken from boolarr
```

```python
x
```

```python
#let update value of 100 to 500 in column 1 to 3 AND row 2 to 5
boolarrr= x[1:6,0:4]==100
boolarrr
```

```python
x[boolarrr,(1:6,0:4)]=500
x
#we can work with specific row and specific column one at tme
#either row and column should be taken as whole
```

```python
#let update value of 100 to 500 in column row 2
boolarrr= x[1,:]==100
boolarrr
```

```python
x[(1),boolarrr]=500
x
```
