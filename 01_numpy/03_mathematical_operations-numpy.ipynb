```python
#logical operations

a=np.random.randint(0,2,5)
b=np.random.randint(0,2,5)
print(a)
print(b)

print(np.logical_or(a,b))
print(np.logical_and(a,b))
print(np.logical_not(a))
```

```python
a=np.random.randint(0,2,(5,4))
b=np.random.randint(0,2,(5,4))
print(a)
print(b)

print(np.logical_or(a,b))
print(np.logical_and(a,b))
print(np.logical_not(a))
```
