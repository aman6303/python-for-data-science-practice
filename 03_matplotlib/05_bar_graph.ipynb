```python
year=[2012,2013,2014,2015,2016,2017]
salary=[12,13,14,17,19,20]
population=[100,120,180,250,300,370]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(year,population)
plt.show()
```

```python
plt.bar(year,population, width=0.5)
plt.show()
```

```python
plt.bar(year,population, width=1)  #by default 0.8
plt.show()
```

```python
#rotating ticks

year=['year-2012','year-2013','year-2014','year-2015','year-2016','year-2017']
plt.bar(year,population, width=0.8)

plt.xticks(rotation=40)  #can be used in ther graphs also
plt.show()
```
