```python
import matplotlib.pyplot as plt
```

```python
year=[2012,2013,2014,2015,2016,2017]
salary=[12,13,14,17,19,20]
population=[100,120,180,250,300,370]
```

```python
plt.scatter(year,salary)
plt.show()
```

```python
plt.scatter(year,salary,s=population,c='r')
plt.title("Bubble Chart")
plt.show()
```

```python
plt.scatter(year,salary,s=population,c='r',alpha=0.7)  #opacity
plt.show()
```

```python
import numpy as np
c=np.arange(len(year))
c[1]=c[0]
plt.scatter(year,salary,s=population,c=c)
plt.show()
```

```python
plt.scatter(year,salary,s=population,c=c,edgecolor='black',marker='^')
plt.title("Bubble Chart")
plt.show()
```

```python
#if u want to add some text

plt.scatter(year,salary,s=population,c=c,edgecolor='black',marker='^')
plt.title("Bubble Chart")

plt.text(year[0]+0.1,salary[0]+0.1,population[0])
plt.show()
```

```python
plt.scatter(year,salary,s=population,c=c,edgecolor='black',marker='^')
plt.title("Bubble Chart")
for i in range(6):
    plt.text(year[i]+0.1,salary[i]+0.1,population[i])
plt.show()
```
