```python
import numpy as np
x=np.arange(0,5,0.1)
y=x**3

#plt.plot(x,y,color='black', marker='o')
plt.plot(x,y,color='black',linewidth=3)
plt.ylabel('x^3')
plt.xlabel('x')
plt.title("demo")
plt.show()
```

```python
import numpy as np
x=np.arange(0,5,0.1)
y=x**3
y2=x*x

plt.plot(x,y,color='black',linewidth=3, label='x^3')
plt.plot(x,y2,color='red',linewidth=3,label='x^2')

plt.xlabel('x')
plt.legend()   #to show label given in plt.plot()
plt.title("demo")
plt.show()
```

```python
import numpy as np
x=np.arange(0,5,0.1)
y=x**3
y2=x*x

plt.plot(x,y,color='black',linewidth=3, label='x^3')
plt.plot(x,y2,color='red',linewidth=3,label='x^2')

plt.axis([0,10,0,150])  #need 4 args x range & y range
plt.xlabel('x')
plt.legend()
plt.title("demo")
plt.show()
```

```python
import numpy as np
x=np.arange(0,5,0.1)
y=x**3
y2=x*x

plt.plot(x,y,color='black',linewidth=3, label='x^3')
plt.plot(x,y2,color='red',linewidth=3,label='x^2')

#plt.axis([0,10,0,150])  #need 4 args x range & y range

plt.ylabel('y')
plt.xlabel('x')

plt.grid()

plt.legend()
plt.title("demo")
plt.show()
```

```python
#adding random text at some points

import numpy as np
x=np.arange(0,5,0.1)
y=x**3
y2=x*x

plt.plot(x,y,color='black',linewidth=3, label='x^3')
plt.plot(x,y2,color='red',linewidth=3,label='x^2')

#plt.axis([0,10,0,150])  #need 4 args x range & y range

plt.ylabel('y')
plt.xlabel('x')

plt.grid()

plt.text(2,80,"text",fontsize=12)  

plt.legend()
plt.title("demo")
plt.show()
```
