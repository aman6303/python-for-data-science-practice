```python
import numpy as np
```

```python
xt=np.arange(23)
```

```python
#creating x ticks

plt.hist(b,edgecolor='black')
#plt.xticks([1,2,3,4,5])
plt.xticks(xt)
plt.show()
```

```python
#giving no of bins

plt.hist(b,bins=25,edgecolor='black')
plt.xticks(xt)
plt.show()
```

```python
plt.hist(b,bins=21,edgecolor='black')
plt.xticks(xt)
plt.show()
```

```python
plt.hist(b,bins=[1,3,6,10,15,20],edgecolor='black')
plt.xticks(xt)

plt.axis([0,25,0,10])
plt.show()
```

```python
plt.hist(b,bins=[1,3,6,10,15,20],edgecolor='black')

plt.axis([0,25,0,10])
plt.show()
```
