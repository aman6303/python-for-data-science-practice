```python
import matplotlib.pyplot as plt
```

```python
sizes=[3,4,6,2]

plt.pie(sizes)
plt.show()
```

```python
labels=['A','B','C','D']

sizes=[3,4,6,2]
colors=['blue','purple','red','pink']
plt.pie(sizes,colors=colors,labels=labels)

plt.title("split among classes")

plt.axis("equal")
plt.show()
```

```python
#calculating percentage autopct

labels=['A','B','C','D']

sizes=[3,4,6,2]
colors=['blue','purple','red','pink']
plt.pie(sizes,colors=colors,labels=labels,autopct='0.2f%%')
        #we can also pass function in autopct to printsomething diff on graph

    plt.title("split among classes")

plt.axis("equal")
plt.show()
```

```python
#highlighting one particular part ---- explode

labels=['A','B','C','D']

sizes=[3,4,6,2]
colors=['blue','purple','red','pink']
explode=[0.1,0.2,0,0]

plt.pie(sizes,colors=colors,explode=explode,labels=labels,autopct='0.2f%%')

plt.title("split among classes")

plt.axis("equal")
plt.show()
```

```python
# making pie clockwise

labels=['A','B','C','D']

sizes=[3,4,6,2]
colors=['blue','purple','red','pink']
explode=[0.1,0.2,0,0]

plt.pie(sizes,colors=colors,explode=explode,labels=labels,autopct='0.2f%%',counterclock=False,startangle=100)

plt.title("split among classes")

plt.axis("equal")
plt.show()
```
