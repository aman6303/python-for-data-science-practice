```python
from selenium import webdriver
from bs4 import BeautifulSoup as bsp
import pandas as pd
from selenium.webdriver.common.by import By
```

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://books.toscrape.com/')
```

```python
a=driver.find_elements(By.CLASS_NAME,'image_container')  #does not select the all block of that class
```

```python
for i in a:
    print(i.html)
```

```python
b=driver.find_elements(By.TAG_NAME,'h3')
```

```python
for i in b:
    print(i)
```

```python
from selenium.common.exceptions import NoSuchElementException

try:
    element = driver.find_element(By.ID, "abc")
    print("Success: Found the element!")
    
except NoSuchElementException:
    # EXCEPT block runs if the element was NOT found
    print("Element not found, but the script is continuing.")
    pass

print("Script finished successfully.")
```
