```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://books.toscrape.com/')
```

```python
# understanding absolute path --(exact path)

driver.find_element(By.XPATH, 'html/head/title')
```

```python
ans = driver.find_element(By.XPATH, 'html/head/title')
```

```python
#getting inner and outer html

print(ans.get_attribute('innerHTML'))   #without current tag
print(ans.get_attribute('outerHTML'))
# print(ans.get_attribute('outerHTML').text)--- give error as we are getting ans in str
```

```python
# finding tag having class="breadcrumb"

ans=driver.find_element(By.XPATH, '//*[@class = "breadcrumb"]')
print(ans.get_attribute('innerHTML'))
print(ans.get_attribute('outerHTML'))
```

```python
#concept of partial x path
```

```python
ans=driver.find_element(By.XPATH, '//div[contains(@class, "h1")]')
print(ans.get_attribute('innerHTML'))
print(ans.get_attribute('outerHTML'))
```

```python
ans=driver.find_element(By.XPATH, '//div[starts-with(@class, "col")]')
print(ans.get_attribute('innerHTML'))
print(ans.get_attribute('outerHTML'))
```

```python
# concepts of predicate
```

```python
ans=driver.find_element(By.XPATH, '//*[contains(@class, "nav")]/li/ul')
print(ans.get_attribute('outerHTML'))
```

```python
ans=driver.find_element(By.XPATH, '//*[contains(@class, "nav")]/li/ul/li[6]')
print(ans.get_attribute('outerHTML'))
```

```python
ans=driver.find_element(By.XPATH, '//*[contains(@class, "nav")]/li/ul/li[last()]')
print(ans.get_attribute('outerHTML'))
```

```python
ans=driver.find_elements(By.XPATH, '//*[contains(@class, "nav")]/li/ul/li[position()<3]')
for i in ans:
    print(i.get_attribute('outerHTML'))
```

# css selector

```python
ans=driver.find_element(By.CSS_SELECTOR, 'html head title')
print(ans.get_attribute('outerHTML'))
```

```python
ans=driver.find_element(By.CSS_SELECTOR, 'html>head>title')
print(ans.get_attribute('outerHTML'))
```

```python
ans=driver.find_element(By.CSS_SELECTOR, 'div li')
print(ans.get_attribute('outerHTML'))
```
