```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
from selenium.webdriver.common.by import By
```

```python
# open bookscrap.com

driver.get('https://books.toscrape.com/')
```

```python
# get music link using element link text

driver.find_element(By.LINK_TEXT, 'Music')
```

```python
mb=driver.find_element(By.LINK_TEXT, 'Music')
mb.click()
```

```python
driver.back()
```

```python
# get link using element partial link text - 'kill'

#driver.find_element(By.PARTIAL_LINK_TEXT, 'kill') 
#error if not link present
```

```python
mb=driver.find_element(By.LINK_TEXT, 'Music')
mb.click()
```

```python
k1=driver.find_element(By.PARTIAL_LINK_TEXT, 'Kill') #return first match
```

```python
k1.click()
```

```python
type(k1) # we get a WebElement type --- methods and properties 
```

```python
driver.back()
```

```python
# to get all match we use find_elements()

driver.find_elements(By.PARTIAL_LINK_TEXT, 'Kill')
```

```python
all_Kill_links=driver.find_elements(By.PARTIAL_LINK_TEXT, 'Kill')

print(all_Kill_links)
```

```python
#go to home page
driver.back()
```

```python
# getting all image tag element

all_img =driver.find_elements(By.TAG_NAME, 'img')
print(len(all_img))
```

```python
# find elements by class name

driver.find_elements(By.CLASS_NAME, "product_pod")
```

```python
all_prd=driver.find_elements(By.CLASS_NAME, "product_pod")
print(len(all_prd))
```

```python
# getting elemets by attribute name-- name='viewport'

all_prd=driver.find_elements(By.NAME, "viewport")
print(len(all_prd))
```
