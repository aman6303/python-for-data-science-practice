```python
# testing driver

from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# 1. Initialize the WebDriver
#    This will open a new Chrome window.
#    If using the automated method, this is all you need.
driver = webdriver.Chrome()

try:
    # 2. Open a website
    driver.get("https://www.google.com")

    # 3. Find an element by its name
    #    (In this case, the Google search bar)
    search_box = driver.find_element(By.NAME, "q")

    # 4. Type text into the element and submit
    search_box.send_keys("Selenium Python")
    search_box.submit()

    # 5. Wait for a moment to see the results
    print("Search successful!")
    time.sleep(5)  # Wait for 5 seconds

finally:
    # 6. Close the browser window
    print("Closing browser.")
    driver.quit()
```
