```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver= webdriver.Chrome()
```

# 1. find title of google images page

```python
driver.get('https://www.google.com/')
```

```python
mage_button=driver.find_element(By.PARTIAL_LINK_TEXT,'mage')
```

```python
mage_button.click()
```

```python
title=driver.find_element(By.XPATH,'html/head/title')
print(title.get_attribute('innerHTML'))
```

```python
title=driver.find_element(By.XPATH,'//title')
print(title.get_attribute('innerHTML'))
```

```python
driver.back()
```
