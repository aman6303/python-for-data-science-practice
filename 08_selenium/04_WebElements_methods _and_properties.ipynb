```python
from selenium import webdriver
```

```python
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://books.toscrape.com/')
```

```python
driver.find_element(By.CLASS_NAME, 'price_color')
```

```python
c1=driver.find_element(By.CLASS_NAME, 'price_color')
print(type(c1))
```

```python
c1.click()  #nothing happens if not clickable
```

```python
c1.clear()  # clear content of elem if it is user editable otherwise give error

# use ful in fillinging and searching
```

```python
# get_attribute(name) --- to get value of attribute
#return none if not available

a1=driver.find_element(By.TAG_NAME,'a')
print(a1.get_attribute('href'))  # we get full url along with base
```

```python
p1=driver.find_element(By.CLASS_NAME,"product_pod")
```

```python
p1.get_attribute('class')
```

```python
# getting tag_name

print(p1.tag_name)
print(a1.tag_name)
print(c1.tag_name)
```

```python
# getting associated text or string
# give complete text in that element 

print(p1.text)
print()
print(a1.text)
print()
print(c1.text)
```
