# 2. getting url of all pages

```python
all_urls=['https://quotes.toscrape.com/page/1/']
base_url='https://quotes.toscrape.com'
current_url='https://quotes.toscrape.com/page/1/'

response=requests.get(current_url)

while response.status_code==200:
    data=BeautifulSoup(response.text,'html.parser')
    l=data.find(class_='next')
    if l==None:
        break
    next_sub=l.a['href']
    next_url=base_url+next_sub
    all_urls.append(next_url)
    print(next_url)
    current_url=next_url
    response=requests.get(current_url)
```

```python
all_authors=[]
for i in all_urls:
    response=requests.get(i)
    data=BeautifulSoup(response.text,'html.parser')
    for j in data.find_all(class_="author"):
        all_authors.append(j.text)
```

```python
all_authors.sort()
for i in all_authors:
    print(i)
```

# 3. printing date of birth

```python
#getting all urls

dob_dict={}

base_url= 'https://quotes.toscrape.com'

for i in all_urls:
    response=requests.get(i)
    data=BeautifulSoup(response.text,'html.parser')
    for j in data.find_all(class_="author"):
        if (j.text).startswith('J'):
            if j.text not in dob_dict.keys():
                author_url=base_url+j.next_element.next_element.next_element['href']
                r=requests.get(author_url)
                data2=BeautifulSoup(r.text,'html.parser')
                dob_dict[j.text] = data2.find(class_="author-born-date").text        
    print(dob_dict)
```

```python
#soritng based on keys

sorted_dict = dict(sorted(dob_dict.items()))
sorted_dict
```
