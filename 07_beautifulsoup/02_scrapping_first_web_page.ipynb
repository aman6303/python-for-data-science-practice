```python
import requests
from bs4 import BeautifulSoup
```

```python
response= requests.get('https://info.cern.ch/hypertext/WWW/TheProject.html')
response
```

```python
response.headers
```

```python
response.text
```

```python
html_data = response.text  #we not need to convert as data is alredy in sr formet (i case of json,  data was json type)
```

```python
#parse

data=BeautifulSoup(html_data,'html.parser')
```

```python
print(data.prettify())
```
