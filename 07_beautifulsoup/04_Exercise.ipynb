# 11. printing the name of all the categories

```python
base_url='https://books.toscrape.com/'
response=requests.get('https://books.toscrape.com/index.html')
print(response.status_code)

data=BeautifulSoup(response.text,'html.parser')

for i in (data.find(class_="nav nav-list")).li.ul.find_all('a'):
    print(i.text.strip())
    print(base_url+i['href'])
```

# 12. print all book names of first 10 pages

```python
first_10_pages_urls=['https://books.toscrape.com/catalogue/page-1.html']

base_url='https://books.toscrape.com/catalogue/'

current_url='https://books.toscrape.com/catalogue/page-1.html'
for i in range(9):
    response=requests.get(current_url)
    data=BeautifulSoup(response.text,'html.parser')
    nex_url=base_url+data.find(class_='next').a['href']
    first_10_pages_urls.append(nex_url)
    current_url=nex_url
print(len(first_10_pages_urls))
```

```python
all_10_pages_name=[]

for i in first_10_pages_urls:
    response=requests.get(i)
    data=BeautifulSoup(response.text,'html.parser')
    for j in data.find_all(class_="product_pod"):
        all_10_pages_name.append(j.h3.a['title'])
```

```python
print(len(all_10_pages_name))
for i in all_10_pages_name:
    print(i)
```

# 13. print detail of first 2 pages books

```python
#geeting urls of books present on first two pages
books_url=[]

for i in range(2):
    response=requests.get(first_10_pages_urls[i])
    data=BeautifulSoup(response.text,'html.parser')
    for j in data.find_all(class_="product_pod"):
        books_url.append(base_url+j.h3.a['href'])
```

```python
import re
books_details=[]

for i in books_url:
    response=requests.get(i)
    data=BeautifulSoup(response.text,'html.parser')
    title=data.h1.string
    link=i
    price=float(re.search('[\d.]+',data.find(class_="price_color").string).group())
    qty=int(re.search('\d+',data.find(class_='instock availability').contents[-1].strip()).group())
    books_details.append([title,link,price,qty])
    print(title)
```

```python
df=pd.DataFrame(books_details,columns=["Name",'Link','Price','Qty'])
df.head()
```

```python
df.to_csv('exercise_books.csv',index=False)
```
