```python
import sqlite3
```

```python
db= sqlite3.connect("Employee.sqlite")
db
```

```python
cur = db.cursor()
cur
```

```python
cur.execute("create table employeedetail (employeeid int primary key, names text, age int, department text, salary int)")
```

```python
values = [(100,"rishav",25,"marketing",50000),(101,"abhishek",26,"operations",60000),(102,"aman",24,"human resource",75000)]

cur.executemany('insert into employeedetail values (?,?,?,?,?)',values)
db.commit()
```

```python
cur.execute("delete from employeedetail where employeeid = 101")
db.commit()
```

```python
# delete a row with parameter
id_del=102

cur.execute("delete from employeedetail where employeeid = ?",(id_del,))
db.commit()
```
