```python
#again establishing the connection

db=sqlite3.connect('school.sqlite') 
db
```

```python
#creating a cursor object which help in writing sql 
cur=db.cursor()
cur
```

```python
cur.execute('insert into student values (103,"Mohit",24)')
db.commit()
```

```python
#use of place holder ?
rollnumber=104
name="Nikhil"
age=34

cur.execute('insert into student values(?,?,?)',(rollnumber,name,age))  #we need to provide values in form of tuple
```

```python
db.commit()
```

```python
values = [(105,"Nidhi",20),(106,"Manisha",21),(107,"Ankush",22)]  #should be list of tuples

cur.executemany('insert into student values(?,?,?)',values)
```

```python
db.commit()
```

```python
#updating

cur.execute('update student set age = 30')
db.commit()
```

```python
# fetching data from sqlite

cur.execute('select * from student')

# we treate cur as iterator 
```

```python
for row in cur:
    print(row)
```

```python
# using fetchone()--return first row

cur.execute('select * from student')
print(cur.fetchone())  #after calling again we get next row
print(cur.fetchone())
```

```python
#fetchmany(var)--return specified var rows

cur.execute('select * from student')
cur.fetchmany(4)   #return in the form of list of tuples
```

```python
cur.execute('select * from student')
cur.fetchmany()   
```

```python
#fetch all
cur.execute('select * from student')
ans=cur.fetchall()   
ans
```

```python
for i in ans:
    print(i) 
```

```python
cur.execute('select * from student where rollnumber > 103')
cur.fetchall()   
```
