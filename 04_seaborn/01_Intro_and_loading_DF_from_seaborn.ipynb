#### Seaborn is very powerful when we have dataset in the form of dataframes

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,1000)

plt.scatter(x,np.sin(x))
plt.show()
```

```python
# if we write sns.set()--- it will get color palat and theme from seaborn

sns.set()

plt.scatter(x,np.sin(x))
plt.show()
```

```python
# just like maatplotlib we need to write plt.show() at  end
```
