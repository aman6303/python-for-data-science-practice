```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()  # setting theme and color pallet of seaborn
```

```python
# loading a dataset from seaborn datasets
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

### Plotting relplot() which is the combination of scatterplot and line plot

```python
sns.relplot(x = 'total_bill', y= 'tip', data = df)
plt.show()
```

```python
# by default it is scatter but we can change its kind

sns.relplot(x = df.total_bill, y = df.tip, kind = 'line')
plt.show()
```

```python
# we can also add third attrinute as hue

sns.relplot(x= df.total_bill, y=df.tip, hue = df.sex)
plt.show()
```

```python
# we can also add another feature by changing style

sns.relplot(x= 'total_bill', y = 'tip', data = df, hue = 'sex', style='day')
plt.show()
```

```python
# if we want to split our graph based on a categorical attribue

sns.relplot(x='total_bill',y='tip',data=df, hue = 'sex', col='day')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df, hue = 'sex', col='time')
plt.show()
```

### Plotting scatterplot()

```python
# similar to above just use scatterplot instead of relplot()

sns.scatterplot(x= df.time, y = df.day)
```

```python
# it also get plotted even after specifying plt.show()
```

### plotting line graph

```python
df = sns.load_dataset('flights')
df.head()
```

```python
sns.relplot(x = 'year',y = 'passengers',data = df, kind= 'line')
```

```python
# using lineplot()

sns.lineplot(x= 'year', y= 'passengers', data= df, hue = 'month')
plt.show()
```

```python
# we can also pass col

sns.relplot(x= 'year', y= 'passengers', data= df,kind='line', col = 'month')
plt.show()
```
