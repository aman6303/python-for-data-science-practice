```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

## Using catplot()

### 1. categorical scatterplots

```python
# stripplot
# in x -- categorical
# in y -- numerical value

sns.catplot(x = df.sex,y=df.total_bill, kind = 'strip')   # by default strip
```

```python
sns.catplot(x = df.sex,y=df.total_bill, kind = 'strip', jitter = 0.01)  #jitter tell spreadness
```

```python
sns.catplot(x = df.sex,y=df.total_bill, kind = 'strip', hue= df.sex)
```

```python
# swarmplot()--- similar to above but avoid overlapping and tell little about distribution

sns.catplot(x = df.sex,y=df.total_bill, kind = 'swarm', hue= df.sex)
```

```python
# we can directly use swarmplot

sns.swarmplot(x = df.sex,y=df.total_bill, hue= df.sex)
```

## 3. categorical estimate plots

```python
df = sns.load_dataset('tips')
```

```python
# -------------------------------
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='day', hue='sex')
plt.title('Count of Customers by Day and Gender')
plt.xlabel('Day of Week')
plt.ylabel('Count')
plt.show()

# -------------------------------
# 2 POINT PLOT
# -------------------------------
plt.figure(figsize=(6,4))
sns.pointplot(data=df, x='day', y='tip', hue='sex', ci='sd', markers='o')
plt.title('Average Tip by Day and Gender')
plt.xlabel('Day of Week')
plt.ylabel('Average Tip ($)')
plt.show()
```
