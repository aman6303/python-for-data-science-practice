# Json conversion to python datatypes

```python
import json
```

```python
#all json data always in string
json_data = '{"Student" : "Mohit"}'
python_data = json.loads(json_data) #json.loads()---convert json to python
```

```python
print(python_data)
print(type(python_data))  #whole json file is converted to dict

print(type(python_data['Student']))  #string of json conv to str
print(python_data['Student'])  #accessing data
```

```python
#eg 2

json_data= '{"rollnumber":101}'  
python_data = json.loads(json_data)
print(python_data['rollnumber'])
print(type(python_data['rollnumber']))  #no of json convert to int or float
```

```python
json_data= '{"rollnumber":101.3}'  
python_data = json.loads(json_data)
print(python_data['rollnumber'])
print(type(python_data['rollnumber'])) 
```

```python
json_data= '{"rollnumber":null}'  
python_data = json.loads(json_data)
print(python_data['rollnumber'])
print(type(python_data['rollnumber'])) #null convert to none
```

```python
json_data= '{"student": {"name":"aman", "roll":101}}'  #object of json
python_data = json.loads(json_data)  #convt to dict of dict

student_details= python_data['student'] #first dict

print(student_details['name'])
print(student_details['roll']) 
```

```python
json_data= '{"fruits": ["apple","banana","orange"]}'  
python_data = json.loads(json_data)

fruits=python_data['fruits']       #array to list 
print(type(fruits))

for i in fruits:
    print(i)
```
